```python
import json
# Load the JSONL file
input_file = "api_test_data.jsonl"
with open(input_file, "r") as f:
    api_test_data = [json.loads(line) for line in f]

# Initialize a list to store results
results = []
```

```python
results = []
for entry in api_test_data:
    stream = client.chat.completions.create(
      model="gpt-4o-mini",
      messages =entry["messages"],
      stream = True,
      temperature=0.7,
      max_tokens=500,
      top_p=1
    )
    
    for chunk in stream:
        if chunk.choices[0].delta.content is not None:
            sentiment = chunk.choices[0].delta.content
        
    # Updated results.append function
    results.append({
        "heading": next((line.replace("Heading: ", "") for line in entry["messages"][1]["content"].split("\n") if line.startswith("Heading: ")), None),
        "pros": next((line.replace("Pros: ", "") for line in entry["messages"][1]["content"].split("\n") if line.startswith("Pros: ")), None),
        "cons": next((line.replace("Cons: ", "") for line in entry["messages"][1]["content"].split("\n") if line.startswith("Cons: ")), None),
        "predicted_sentiment": sentiment
    })

    
# Convert results to a DataFrame
results_df = pd.DataFrame(results)
```

```python
import pandas as pd
import json
from sklearn.metrics import classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load the results CSV (GPT-4 predictions)
# results_csv_path = "api_test_results.csv"  # Leave these lines commented out
# results_df = pd.read_csv(results_csv_path)

# Load the JSONL file (actual sentiments)
jsonl_path = "api_test_data_with_sentiments.jsonl"
with open(jsonl_path, 'r') as f:
    actual_sentiments_data = [json.loads(line) for line in f]

# Extract actual sentiments from the JSONL file
actual_sentiments = [entry['sentiment'] for entry in actual_sentiments_data]

# Add the actual sentiments to the results DataFrame
results_df['actual_sentiment'] = actual_sentiments

# Normalize the predicted_sentiment and actual_sentiment columns
results_df['predicted_sentiment'] = results_df['predicted_sentiment'].str.lower().str.strip()
results_df['actual_sentiment'] = results_df['actual_sentiment'].str.lower().str.strip()

# Ensure only valid classes remain
valid_classes = ['positive', 'neutral', 'negative']
results_df = results_df[
    results_df['actual_sentiment'].isin(valid_classes) &
    results_df['predicted_sentiment'].isin(valid_classes)
]

# Evaluate GPT-4 predictions with classification metrics
classification_report_output = classification_report(
    results_df['actual_sentiment'],
    results_df['predicted_sentiment'],
    target_names=['positive', 'neutral', 'negative']
)

# Confusion Matrix
conf_matrix = confusion_matrix(
    results_df['actual_sentiment'],
    results_df['predicted_sentiment'],
    labels=['positive', 'neutral', 'negative']
)

# Print Classification Report
print("\nClassification Report:")
print(classification_report_output)

# Visualize Confusion Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['positive', 'neutral', 'negative'], yticklabels=['positive', 'neutral', 'negative'])
plt.title("Confusion Matrix for GPT-4 Predictions")
plt.xlabel("Predicted Sentiment")
plt.ylabel("Actual Sentiment")
plt.show()
```
